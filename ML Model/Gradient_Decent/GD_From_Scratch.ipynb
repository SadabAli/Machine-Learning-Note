```python
X,y = make_regression(n_samples=100 , n_features=1 , n_informative=1, n_targets=1,noise=20,random_state=32)
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
print("Slope m",lr.coef_)
print("Intercept b",lr.intercept_)
```

Slope "m" is constand and b is variable

```python
class GDalgorithm:

    def __init__(self,learning_rate , epochs):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.m = 38.9817587
        self.b=-242
    
    def fit(self,X,y):
        # calculating the intercept "b" using GD
        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y-self.m * X.ravel() - self.b)
            self.b = self.b - (self.learning_rate * loss_slope)
            print(loss_slope,self.b)
        print(self.b) 
```

```python
gd = GDalgorithm(0.001,100)
```
