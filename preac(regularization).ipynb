```python
import pandas as pd 
import numpy as np 

from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression , LogisticRegression,SGDClassifier,Ridge,Lasso,ElasticNet
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score 
```

```python
df = pd.read_csv(r"C:\Users\alisa\OneDrive\Desktop\Machine Learning Note\Malware dataset.csv")
df.sample(2)
```

```python
df.drop(columns=['hash'],inplace=True)
```

```python
df.sample(2)
```

```python
df['classification'] = df.classification.map({'benign':0, 'malware':1})
```

```python
df.shape
```

```python
df = df.sample(100000)
df.head(3)
```

```python
X = df.drop(columns=['classification'])
y=df['classification']
```

```python
print(X.sample(2))
print(y.sample(2))
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=.3,random_state=42)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train,y_train)
linear_reg_pred = linear_reg.predict(X_test)
print(r2_score(y_test , linear_reg_pred))
print(cross_val_score(linear_reg,X,y,cv=10).mean())
```

```python
logist_reg = LogisticRegression()
logist_reg.fit(X_train,y_train)
logist_reg_pred = logist_reg.predict(X_test)
print(r2_score(y_test , logist_reg_pred))
print(cross_val_score(logist_reg,X,y,cv=10))
```

```python
rid = Ridge(alpha=0.01,max_iter=100)
rid.fit(X_train,y_train)
rid_pred = rid.predict(X_test)
print(r2_score(y_test , rid_pred))
print(cross_val_score(rid,X,y,cv=10).mean())
```

```python
lass = Lasso(alpha=0.01,max_iter=1000)
lass.fit(X_train,y_train)
lass_pred = lass.predict(X_test)
print(r2_score(y_test , rid_pred))
print(cross_val_score(lass,X,y,cv=10))
```

```python
elas = ElasticNet(alpha=0.01,l1_ratio=0.8)
elas.fit(X_train,y_train)
elas_pred = elas.predict(X_test)
print(r2_score(y_test ,elas_pred))
print(cross_val_score(elas,X,y,cv=10))
```
