```python
import pandas as pd 
import numpy as np 
from sklearn.datasets import make_classification 
from sklearn.model_selection import train_test_split , cross_val_score 
from sklearn.ensemble import BaggingClassifier  , RandomForestClassifier 
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

Bagging me row and col. span lagta hai bass root me (tree like cols. or row span) WHERE AS Random forest me node wise spanning lagta hai

```python
X,y = make_classification(n_samples=100 , n_features=5,n_informative=5,n_clusters_per_class=1,n_redundant=0)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

```python
bag  = BaggingClassifier(max_features=2)# cols span
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python

rf = RandomForestClassifier(max_features=2)
     

rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
```
